```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("cinema_hall_ticket_sales.csv")

# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()

# Display dataset info
print("Dataset Info:\n", df.info())
print("\nFirst 5 Rows:\n", df.head())

# Check column names
print("\nColumn Names:", df.columns.tolist())

# Summary statistics
print("\nSummary Statistics:\n", df.describe())

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Data Visualization

# Histogram of Ticket Prices
if "Ticket_Price" in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df["Ticket_Price"], bins=20, kde=True)
    plt.title("Distribution of Ticket Prices")
    plt.xlabel("Ticket Price")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Column 'Ticket_Price' not found in dataset.")

# Bar Chart for Movie Genres
if "Movie_Genre" in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x="Movie_Genre", order=df["Movie_Genre"].value_counts().index)
    plt.xticks(rotation=45)
    plt.title("Number of Tickets Sold per Genre")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.show()
else:
    print("Column 'Movie_Genre' not found in dataset.")

# Box Plot of Ticket Prices
if "Ticket_Price" in df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, y="Ticket_Price")
    plt.title("Box Plot of Ticket Prices")
    plt.ylabel("Price")
    plt.show()
else:
    print("Column 'Ticket_Price' not found in dataset.")

# Count Plot for Seat Type
if "Seat_Type" in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x="Seat_Type", order=df["Seat_Type"].value_counts().index)
    plt.title("Distribution of Seat Types")
    plt.xlabel("Seat Type")
    plt.ylabel("Count")
    plt.show()
else:
    print("Column 'Seat_Type' not found in dataset.")

# Count Plot for Purchase Again
if "Purchase_Again" in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x="Purchase_Again")
    plt.title("Customer Willingness to Purchase Again")
    plt.xlabel("Purchase Again (Yes/No)")
    plt.ylabel("Count")
    plt.show()
else:
    print("Column 'Purchase_Again' not found in dataset.")
```
